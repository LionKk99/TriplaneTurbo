```python
import plotly.graph_objects as go
import numpy as np
import re

def load_obj_file(obj_path):
    """ .obj """
    vertices = []
    faces = []
    
    with open(obj_path, 'r') as f:
        for line in f:
            line = line.strip()
            if line.startswith('v '):
                #  (x, y, z, r, g, b)
                parts = line.split()[1:]
                if len(parts) >= 6:
                    vertex = [float(x) for x in parts[:3]]  # x, y, z
                    color = [float(x) for x in parts[3:6]]  # r, g, b (0-1)
                    vertices.append(vertex)
            elif line.startswith('f '):
                # 
                parts = line.split()[1:]
                face = []
                for part in parts:
                    #  "1/1/1"
                    vertex_idx = int(part.split('/')[0]) - 1  # obj 1
                    face.append(vertex_idx)
                if len(face) >= 3:
                    faces.append(face)
    
    return np.array(vertices), faces

def visualize_obj(obj_path):
    """ plotly  .obj """
    vertices, faces = load_obj_file(obj_path)
    
    #  3D 
    if faces:
        fig = go.Figure(data=[
            go.Mesh3d(
                x=vertices[:, 0],
                y=vertices[:, 1], 
                z=vertices[:, 2],
                i=[f[0] for f in faces],
                j=[f[1] for f in faces],
                k=[f[2] for f in faces],
                colorscale='Viridis',
                opacity=0.8,
                lighting=dict(
                    ambient=0.3,
                    diffuse=0.8,
                    specular=0.1
                )
            )
        ])
        
        fig.update_layout(
            title="3D Model Visualization",
            scene=dict(
                xaxis_title='X',
                yaxis_title='Y', 
                zaxis_title='Z',
                aspectmode='data'
            ),
            width=800,
            height=600
        )
        
        fig.show()
    else:
        print("No faces found in the obj file")

# 
obj_path = "/data2/hja/3D/TriplaneTurbo/outputs/TriplaneTurbo_v1/dreamfusion_415_prompt_library/save/it0-export/__Luminescent_wild_horses/model.obj"
visualize_obj(obj_path)
```
